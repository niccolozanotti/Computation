```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc
from matplotlib import rc
import scienceplots
rc('text', usetex=True)
# style used in scientific papers(LaTeX based)
plt.style.use(['science'])
```

```python
x = np.linspace(0, 50, 10000)
J1 = sc.jvp(1,x)
J2 = sc.jvp(2,x)
J3 = sc.jvp(3,x)
J4 = sc.jvp(4,x)
```

```python
fig, axs = plt.subplots()

l1, = axs.plot(x,J1,linestyle='solid',linewidth=1.2)
l2, = axs.plot(x,J2,linestyle='solid',linewidth=1.2)
l3, = axs.plot(x,J3,linestyle='solid',linewidth=1.2)
l4, = axs.plot(x,J4,linestyle='solid',linewidth=1.2)

axs.set_xlim(0,17,100)
# axs.set_ylim(-0.4,1)
axs.set_title(r'Bessel functions of the 1st Kind')
axs.set_ylabel('$J_n(x)$',fontsize=8)
axs.set_xlabel('$x$',fontsize=8)

# axs.grid(visible=True, which='major', color='.9', linestyle='-')
# axs.grid(visible=True, which='minor', color='gray', linestyle='dotted', alpha=0.2)
axs.minorticks_on()


axs.legend((l1,l2,l3,l4),
           (r'$J_1$','$J_2$','$J_3$','$J_4$'),
           loc='best',
           ncol=1,
           fontsize=6)
# axs.text(550,.005,
#          '$I_0 = (2.92 \pm 3.33) \cdot 10^{-6}$ mA\n$\eta V_T = (50.66 \pm 5.05)$ mV',
#          fontsize = 6,
#          bbox = dict(boxstyle="round,pad=0.6",facecolor = 'palegoldenrod', alpha = 0.5,edgecolor='None'))
# plt.savefig('../figs/I-session/silicon-i-v-semilog.pdf')
plt.savefig('bessel-functions-1stkind.pdf')
```
